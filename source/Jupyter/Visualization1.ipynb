```python
import os
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] # 显示中文
plt.rcParams['axes.unicode_minus'] = False # 显示负号
```

```python
df=pd.read_csv(
    r'E:\Projects\OneDrive - somebottle\文档Documents\大数据分析Spark\期末大作业\analyses\pollutant_average\part-00000-e3d62509-406d-44a3-86a3-a4acea2bf1ee-c000.csv'
)
```

```python
# 指定x轴数据标签
labels = ['SO2', 'NO2', 'PM10', 'PM2.5', 'O3', 'CO']
# 创建柱状图
plt.bar(labels, df.iloc[0])
# 设置图表标题和标签
plt.title('各污染物平均浓度')
plt.xlabel('污染物')
plt.ylabel('浓度 (μg/m3 or mg/m3)')
# 显示图表
plt.show()
```
